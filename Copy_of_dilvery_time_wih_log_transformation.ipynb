```python
# transforming variables for accuracy
```

```python
model1 = smf.ols("D_T~np.log(S_T)",data = df2).fit()
```

```python
model1.params
```

```python
model1.summary()
```

```python
print(model1.conf_int(0.01))
```

```python
pred1 = model1.predict(pd.DataFrame(df2['S_T']))
```

```python
pred1.corr(df2.D_T)
```

```python
pred1
```

```python
plt.scatter(x=df2['S_T'],y=df2['D_T'],color='green');plt.plot(df2['S_T'],pred1,color='blue');plt.xlabel('S_T');plt.ylabel('D_T')
```

```python
# Exponential transformation
model2 = smf.ols('np.log(D_T)~S_T',data=df2).fit()
model2.params
model2.summary()
print(model2.conf_int(0.01)) # 99% confidence level
pred_log = model2.predict(pd.DataFrame(df2['S_T']))
pred_log
pred2=np.exp(pred_log)  # as we have used log(AT) in preparing model so we need to convert it back
pred2
pred2.corr(df2.D_T)
plt.scatter(x=df2['S_T'],y=df2['D_T'],color='green');plt.plot(df2.S_T,np.exp(pred_log),color='blue');plt.xlabel('S_T');plt.ylabel('D_T')
resid_2 = pred2-df2.D_T
```

```python
# so we will consider the model having highest R-Squared value which is the log transformation - model2
# getting residuals of the entire data set
student_resid = model2.resid_pearson
student_resid
plt.plot(model2.resid_pearson,'o');plt.axhline(y=0,color='green');plt.xlabel("Observation Number");plt.ylabel("Standardized Residual")

# Predicted vs actual values
plt.scatter(x=pred2,y=df2.D_T);plt.xlabel("Predicted");plt.ylabel("Actual")
```
