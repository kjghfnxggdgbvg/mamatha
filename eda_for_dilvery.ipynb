```python
df1=df.rename({'Delivery Time':'D_T'},axis=1)
```

```python
df1
```

```python
df2=df1.rename({'Sorting Time':'S_T'},axis=1)
```

```python
df2
```

```python
# fitting the linear regression model
```

```python
import statsmodels.formula.api as smf
```

```python
model = smf.ols("D_T~S_T",data=df2).fit()
```

```python
sns.regplot(x="S_T", y="D_T", data=df2 ,color='mediumorchid');
```

```python
model.summary()
```

```python
# here rsquared value is 68% i,e:  based on only sorting time cant say delivery time
```

```python
model.params
```

```python
print(model.tvalues)
```

```python
print(model.pvalues)
```

```python
(model.rsquared,model.rsquared_adj)
```
